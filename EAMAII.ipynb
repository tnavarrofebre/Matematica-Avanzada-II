$$f(z)=\frac{e^{zt}}{\sqrt{z}};\;\;\;con:\;t=\frac{1}{10}$$

con $z=re^{i\theta}$ evaluada entre:

$$z:1\leq r \leq 16\;\;y\;\; -2\pi \leq \theta \leq 2\pi$$

```python
%matplotlib notebook
%pylab
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as col

  
import matplotlib.cm as cm 

   
# compute data to plot 
r, theta = np.mgrid[1:15:100j, -2*np.pi:2*np.pi:100j] 
z = r * np.exp(1j*theta)  
w = 1/(np.sqrt(r)* np.exp(1j*theta/2))*(np.exp(z*.1))
 
# plot data  
fig = plt.figure()  
z_data, c_data = w.real, w.imag 
z_comp, c_comp = 'Re', 'Im' 
c_data = (c_data - c_data.min()) / c_data.ptp() 
colors = cm.hsv((c_data)*1.3) 

ax = fig.add_subplot(projection='3d') 


surf=ax.plot_surface(z.real, z.imag, z_data, facecolors=colors,
                       clim=[z_data.min(), z_data.max()])
ax.set_xlabel('$Re z$')  
ax.set_ylabel('$Im z$')   
ax.set_zlabel(f'${z_comp} w$')  

norm = col.Normalize(vmin=z_data.min(), vmax=z_data.max())

plt.colorbar(cm.ScalarMappable(norm=norm, cmap=cm.hsv),label=f'${c_comp} w$')


plt.show()
```
