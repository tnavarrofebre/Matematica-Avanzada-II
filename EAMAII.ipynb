<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <b>1) $$y''-2y'-4y=3\sin(x)\;\;\;y(0)=1;\;y'(0)=0$$</b> 
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Se aplica transformada de Laplace miembro a miembro,
    $$\mathscr{L}\left\{ y'' \right\} - 3\mathscr{L}\left\{ y' \right\} - 4\mathscr{L}\left\{ y \right\}=3\mathscr{L}\left\{\sin(x) \right\}$$
    de lo que se obtiene,
    $$s^2\mathscr{L}\left\{ y \right\}-sy_0-y'_0-3\left[s\mathscr{L}\left\{ y \right\}-y_0\right]-4\mathscr{L}\left\{ y \right\}=3\mathscr{L}\left\{\sin(x) \right\}$$
    finalmente despejando y sacando factor comun,
    $$\tilde y(s^2-3s-4)=\frac{3}{s^2+1}+s-3$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    La transformada obtenida resulta:
    $$\tilde y=\frac{s^3-3s^2+s}{(s^2+1)(s-4)(s+1)}$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Se realiza la descomposicion en fracciones parciales de acuerdo a,
    $$\frac{As+D}{s^2+1}+\frac{B}{s-4}+\frac{C}{s+1}=\frac{s^3-3s^2+s}{(s^2+1)(s-4)(s+1)}$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    de lo que se obtiene el siguiente sistema de ecuaciones,
    $$\left\{\matrix{A+B+C=1 \\ -3A+D+B-4C=-3 \\ -3D-4A+B+C=1 \\ -4D+B-4C=0}\right.$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    obteniendo:
    $$\matrix{A=\frac{9}{34} \\ B=\frac{4}{17} \\ C=\frac{1}{2} \\ D=-\frac{15}{34}}$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Reemplazando los parametros en la ecuacion,
    $$\tilde y= \frac{9S-15}{34(s^2+1)}+\frac{4}{17(s-4)}+\frac{1}{2(s+1)}$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    luego por tabla se obtiene:
    $$y=\frac{9}{34}\cos(x)-\frac{15}{34}\sin(x)+\frac{4}{17}e^{4x}+\frac{1}{2}e^{-x}$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <b>2) $$y''+y'-2y=3\cos(3x)-11\sin(3x)\;\;\;y(0)=1;\;y'(0)=6$$</b> 
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>

    $$\mathscr{L}\left\{ y'' \right\} +\mathscr{L}\left\{ y' \right\} - 2\mathscr{L}\left\{ y \right\}=3\mathscr{L}\left\{\cos(3x) \right\}-11\mathscr{L}\left\{\sin(3x) \right\}$$
    de lo que se obtiene,
    $$s^2\mathscr{L}\left\{ y \right\}-sy_0-y'_0+\left[s\mathscr{L}\left\{ y \right\}-y_0\right]-2\mathscr{L}\left\{ y \right\}=3\mathscr{L}\left\{\cos(3x) \right\}-11\mathscr{L}\left\{\sin(3x) \right\}$$
    finalmente despejando y sacando factor comun,
    $$\tilde y(s^2+s-2)=\frac{3s-33}{s^2+3^2}+s+7$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>

    $$\tilde y=\frac{s^3+9s^2+12s+48}{(s^2+9)(s-1)(s+2)}$$

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>

</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <b>3) Use la integral de Bromwich para calcul√±ar la transformada de Laplace inversa de:</b><br>
    $$\tilde{f}(s)=\frac{1}{\sqrt{s}}$$
    <br><br>
    La funcion $\tilde{f}(s)$ posee dos ramas con periodicidad $4\pi$ y un corte de rama sobre el eje de los reales negativos desde el punto de rama $\mathcal S= 0+i0$ en el cual la funcion no esta definida. Dadas estas caracteristicas cualquier camino integral debe excluir el corte de rama y el punto de rama.
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    La inversa de Laplace es calculada sobre $\tilde{f}(s)$ de modo que,
    $$\mathscr{L}^{-1}\left\{\tilde{f}\right\}(t)=f(t)$$<br>
    $$f(t)=\frac{1}{2\pi i}\lim_{\delta\to\infty}\int_{\beta-i\delta}^{\beta+i\delta}\tilde{f}(s)e^{st}ds$$
    luego la anti trasformada existe cuando la integral sobre $s$ resulta convergente y la funcion $\tilde{f}(s)$ tiene al menos una discontinuidad en el espacio sobre el que se integra, de otra forma la solucion es identicamente nula. Asi, $\beta$ debe seleccionarse de tal manera que toda discontinuidad quede a derecha o izquierda de la recta de integracion e integrar considerando dichas restricciones.
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    El punto de rama en el origen y la discontinuidad hacia los reales negativos nos obliga a definir $\beta$ como,
    $$\beta\geq\lim_{\alpha\to 0^+}\alpha$$
    quedando la integral como sigue:
    $$f(t)=\frac{1}{2\pi i}\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{\beta-i\delta}^{\beta+i\delta}s^{-\frac{1}{2}}e^{st}ds$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    pero la presencia del corte de rama exige incorporar dos semirectas sobre el eje real negativo y paralelas a este desde $-\infty$ hasta $\mathcal{S}$, una por encima y otra por debajo del mismo con una distancia en el eje imaginario del dominio arbitrariamente peque√±a pero no nula, y una semicircunferencia de radio infinitecimal sobre el semiplano positivo del dominio evitando a $\mathcal{S}$, conectando asi ambas rectas de integracion, finalmente se cierra el contorno $\Gamma$ tal como muestra la figura. Dada la ausencia de discontinuidades dentro del camino de integracion la suma de todas las integrales sobre estos caminos es nula. 
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    $$\frac{1}{2\pi i}\oint_{\Gamma} s^{-\frac{1}{2}}e^{st}ds\stackrel{TIC}{=}0\;\;\hbox{TIC: Teorema Integral de Cauchy.}$$
    $$\frac{1}{2\pi i}\left[\int_{\Gamma_1} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_2} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_3} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_4} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_5} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_6} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_7} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_8} s^{-\frac{1}{2}}e^{st}ds\right]=0$$
    $$\frac{1}{2\pi i}\int_{\Gamma_1} s^{-\frac{1}{2}}e^{st}ds=\frac{1}{2\pi i}\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{\beta-i\delta}^{\beta+i\delta}s^{-\frac{1}{2}}e^{st}ds$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    lo que permite reescribir $f(t)$ como:
    $$f(t)=-\frac{1}{2\pi i}\left[\int_{\Gamma_2} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_3} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_4} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_5} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_6} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_7} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_8} s^{-\frac{1}{2}}e^{st}ds\right]$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Se resuleve cada integral como sigue:
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <i>Integral sobre el camino $\Gamma_2$:</i><br><br>
    Se utiliza la desigualdad ML (forma particular del Teorema de Jordan) tomando como medida del intervalo $L=\beta$ de forma que si existe un $M$ tal que:
    $$M=\max_{x\in[0,\beta]}\lim_{\delta\to\infty}\left|\frac{e^{(x+i\delta)t}}{\sqrt{x+i\delta}}\right|$$
    entonces:
    $$\int_{\Gamma_2} s^{-\frac{1}{2}}e^{st}ds\leq ML$$<br><br>
    $$\max_{x\in[0,\beta]}\lim_{\delta\to\infty}\left|\frac{e^{(x+i\delta)t}}{\sqrt{x+i\delta}}\right|=\lim_{\delta\to\infty}\left|\frac{e^{(\beta+i\delta)t}}{\sqrt{\beta+i\delta}}\right|$$
    $$\lim_{\delta\to\infty}\left|\frac{e^{(\beta+i\delta)t}}{\sqrt{\beta+i\delta}}\right|\leq\lim_{\delta\to\infty}\frac{e^{\beta t}}{\sqrt{\delta}}$$
    $$\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\frac{e^{\beta t}}{\sqrt{\delta}}=0$$<br><br>
    finalmente,
    $$\int_{\Gamma_2} s^{-\frac{1}{2}}e^{st}ds=0$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    De manera similar puede probarse que el valor de la <i>integral sobre el camino $\Gamma_8$</i> resulta cero,
    $$\int_{\Gamma_8} s^{-\frac{1}{2}}e^{st}ds=0$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <i>Integral sobre el camino $\Gamma_3$:</i><br><br>
    Teniendo en cuenta que $s$ puede escribirse en su forma cartesiana bajo una transformacion polar,
    $$s=x+iy \\ x=-\delta\cos(\theta) \;\; y=\delta\sin(\theta) \\ |J|=\delta$$<br>
    Se tiene,
    $$\left|e^{st}\right|=e^{-\delta t\cos(\theta)}$$
    $$\left|s^{-\frac{1}{2}}\right|=\delta^{-\frac{1}{2}}$$<br>
    luego:
    $$s^{-\frac{1}{2}}e^{st}ds=\delta^{\frac{1}{2}}e^{-\delta t\cos(\theta)}d\theta$$
    $$\lim_{\delta\to\infty}\delta^{\frac{1}{2}}e^{-\delta t\cos(\theta)}=0$$<br>
    finalmente,
    $$\int_{\Gamma_3} s^{-\frac{1}{2}}e^{st}ds=0$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    De manera similar puede probarse que el valor de la <i>integral sobre el camino $\Gamma_7$</i> resulta cero,
    $$\int_{\Gamma_7} s^{-\frac{1}{2}}e^{st}ds=0$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <i>Integral sobre el camino $\Gamma_5$:</i><br><br>
    Teniendo en cuenta que $s$ puede escribirse en su forma polar,
    $$s=re^{i\theta} \\ ds=ire^{i\theta}d\theta \\ s^{-\frac{1}{2}}=r^{-\frac{1}{2}}e^{-i\frac{\theta}{2}}$$<br>
    y considerando a $r=\beta$ se obtiene la integral:<br>
    $$\int_{\Gamma_5} s^{-\frac{1}{2}}e^{st}ds=i\lim_{\beta\to 0^+}\int_{\frac{\pi}{2}}^0 \beta^{\frac{1}{2}}e^{i\frac{\theta}{2}}e^{t\beta e^{i\theta}}d\theta$$
    Dado que $\beta$ no depende de la variable de integracion se puede tomar el limite dentro de la integral,
    $$i\int_{\frac{\pi}{2}}^0 e^{i\frac{\theta}{2}}\lim_{\beta\to 0^+}\beta^{\frac{1}{2}}e^{t\beta e^{i\theta}}d\theta$$
    luego,
    $$\lim_{\beta\to 0^+}\beta^{\frac{1}{2}}e^{\beta te^{i\theta}}$$<br>
    Se sabe que $te^{i\theta}$ esta acotado a $|t|$ entonces:
    $$\lim_{\beta\to 0^+}\beta^{\frac{1}{2}}e^{\beta te^{i\theta}}=0$$
    finalmente,
    $$\int_{\Gamma_5} s^{-\frac{1}{2}}e^{st}ds=0$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Ahora podemos reescribir $f(t)$ como:
    $$f(t)=-\frac{1}{2\pi i}\left[\int_{\Gamma_4} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_6} s^{-\frac{1}{2}}e^{st}ds\right]$$
    la cual es una integral en variable real dado que es constante para los valores imaginarios. 
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Ahora podemos reescribir $f(t)$ como:
    $$f(t)=-\frac{1}{2\pi i}\left[\int_{\Gamma_4} s^{-\frac{1}{2}}e^{st}ds+\int_{\Gamma_6} s^{-\frac{1}{2}}e^{st}ds\right]$$
    la cual es una integral en variable real dado que es constante para los valores imaginarios.<br><br>
    $$\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{-\delta}^{\beta}(-x)^{-\frac{1}{2}}e^{-i\pi/2}e^{tx}dx+\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{\beta}^{-\delta}(-x)^{-\frac{1}{2}}e^{i\pi/2}e^{tx}dx$$<br>
    $$-i\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{\beta}^{\delta}x^{-\frac{1}{2}}e^{-tx}dx-i\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{\beta}^{\delta}x^{-\frac{1}{2}}e^{-tx}dx=-2i\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{\beta}^{\delta}x^{-\frac{1}{2}}e^{-tx}dx$$<br>
    $$-2i\lim_{\matrix{\delta\to\infty \\ \beta\to 0^+}}\int_{\beta}^{\delta}x^{-\frac{1}{2}}e^{-tx}dx=-2it^{-\frac{1}{2}}\Gamma\left(\frac{1}{2}\right)$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Finalmente
    $$f(t)=-\frac{1}{2\pi i}\left[-2it^{-\frac{1}{2}}\Gamma\left(\frac{1}{2}\right)\right]$$
    entonces,<br><br>
    $$f(t)=\frac{1}{\sqrt{\pi t}}$$
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <b>4)  Resuelva la ecuaci√≥n diferencial</b><br>
    $$\partial_xU+\partial_tU=x,\;\;x>0,\;t>0$$
    Condiciones iniciales: $U(0,t)=0,\;t>0$ y $U(x,0)=0,\;x>0$.
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    En principio se debe notar la naturaleza en derivadas parciales de la ecuacion diferencial, porl lo tanto se transforma con respecto a $t$ y se toma la variable $x$ como un parametro de la funcion, asi:
    $$\partial_x\mathscr{L}\left\{ U(x,t) \right\}+\mathscr{L}\left\{ \partial_tU(x,t) \right\}=\mathscr{L}\left\{ x \right\}$$
    $$\partial_x\mathscr{L}\left\{ U(x,t) \right\}+s\mathscr{L}\left\{ U(x,t) \right\}-U(x,0)=x\mathscr{L}\left\{ 1 \right\}$$
    de lo que se obtiene una ecuacion diferencial de primer orden no homogenea.
    $$\partial_xU(x,s)+sV(x,s)=\frac{x}{s}$$
    La solucion a la ecuacion anterior sera la suma de la solucion a la ecuacion homogenea $V_h$ y otra particular $V_p$ dada por la inhomogeneidad de la ecuacion.
    $$V_h=Ae^{-sx}$$
    Luego $V_p$ puede obteners epor el metodo de coeficientes indeterminados
   proponiendo la solucion:
    $$V_p=C_1x+C_2\;\;V_p'=C_1$$
    sustituyendo en la ecuacion  y agupando terminos:
    $$xC_1s+(sC_2+C_1)=\frac{1}{s}x$$
    igualando termino a termino:
    $$\left\{\matrix{C_1s=\frac{1}{s} \\ sC_2+C_1=0}\right.$$
    obteniendo valores para las constantes:
    $$C_1=\frac{1}{s^2}\;\;C_2=-\frac{1}{s^3}$$
    De lo anterior la superposicion de soluciones sera:
    $$V=V_h+V_p$$
    $$V=Ae^{-sx}+\frac{1}{s^2}x-\frac{1}{s^3}$$
    finalmente aplicando las condiciones iniciales:
    $$V(x,s)=\int_0^\infty U(x,t)e^{-st}dt$$
    $$V(0,s)=\int_0^\infty 0e^{-st}dt=0$$
    de aqui se obtiene:
    $$A=\frac{1}{s^3}$$
    Se puede, entonces, escribir la solucion general $V(x,s)$, como sigue:
    $$V(x,s)=\frac{1}{s^3}e^{-sx}+\frac{1}{s^2}x-\frac{1}{s^3}$$
    <br><br>
    Solo queda aplicar la antitransformada de Laplace:
    $$\mathscr{L}^{-1}\left\{ V \right\}(x,t)=\mathscr{L}^{-1}\left\{\frac{1}{s^3}e^{-sx}\right\}+\mathscr{L}^{-1}\left\{\frac{1}{s^2}x\right\}-\mathscr{L}^{-1}\left\{\frac{1}{s^3}\right\}$$
    para esto se debe utilizar las propiedades de traslacion y transformada de un monomio,<br>
    $$\hbox{Traslacion:}\;\;\mathscr{L}\left\{ F(t-b) \right\}=e^{-bs}f(s)$$
    $$\hbox{Transformada de un monomio:}\;\;\mathscr{L}\left\{ t^n \right\}=\frac{n!}{s^{n+1}}$$ 
    $$\mathscr{L}^{-1}\left\{ V \right\}(x,t)=\frac{1}{2!}\mathscr{L}^{-1}\left\{\frac{2!}{s^3}e^{-sx}\right\}+\frac{x}{1!}\mathscr{L}^{-1}\left\{\frac{1!}{s^2}\right\}-\frac{1}{2!}\mathscr{L}^{-1}\left\{\frac{2!}{s^3}\right\}$$
    entonces $U(x,t)$ resulta:
    <br><br>
    $$U(x,t)=\frac{1}{2}(t-x)^2+xt-\frac{1}{2}t^2$$
    $$U(x,t)=\frac{1}{2}t^2+\frac{1}{2}x^2-xt+xt-\frac{1}{2}t^2$$
    $$U(x,t)=\frac{1}{2}x^2$$
 </div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <b>5)  Resuelva la ecuaci√≥n diferencial</b><br>
    $$\partial_tU-\partial_{xx}U=0,\;\;0<x<2,\;t>0$$
    Condiciones iniciales: $U(0,t)=0,\;U(2,t)=0,$ y $U(x,0)=3\sin(2\pi x)$.
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Al igual que en el ejercicio $-4-$ notamos la naturaleza en derivadas parciales de la ecuacion diferencial, por lo que transformamos con respecto a $t$ y se toma la variable $x$ como un parametro de la funcion, quedando:
    $$\partial_{xx}\mathscr{L}\left\{ U(x,t) \right\}=\mathscr{L}\left\{ \partial_tU(x,t) \right\}$$
    $$s\mathscr{L}\left\{ U(x,t) \right\}-U(x,0)=\partial_{xx}\mathscr{L}\left\{ U(x,t) \right\}$$
    Donde, haciendo uso de la condicion $U(x,0)=3\sin(2\pi x)$ se obtiene una ecuacion diferencial de primer orden no homogenea:
    $$\partial_{xx}V(x,s)-sV(x,s)=-3\sin(2\pi x)$$
    La solucion a la ecuacion anterior sera la suma de las solucions entre la ecuacion homogenea $V_h$ y la particular $V_p$.
    $$V_h=Ae^{-sx}+Be^{sx}$$
    Luego $V_p$ puede obtenerse haciendo uso de coeficientes indeterminados proponiendo la solucion:
    $$V_p=C_1\cos(2\pi x)+C_2\sin(2\pi x) \\ V_p'=-C_12\pi\sin(2\pi x)+C_22\pi\cos(2\pi x) \\ V_p''=-C_14\pi^2\cos(2\pi x)-C_24\pi^2\sin(2\pi x)$$
    sustituyendo en la ecuacion  y agupando terminos:
    $$\cos(2\pi x)(-4\pi^2C_1-sC_1)+\sin(2\pi x)(-4\pi^2C_2-sC_2)=-3\sin(2\pi x)$$
    igualando termino a termino:
    $$\left\{\matrix{-4\pi^2C_1-sC_1=0 \\ -4\pi^2C_2-sC_2=-3}\right.$$
    obteniendo valores para las constantes:
    $$C_1=0\;\;C_2=\frac{3}{4\pi^2+s}$$
    De l anterior la superposicion de soluciones sera:
    $$V=V_h+V_p$$
    $$V=Ae^{-sx}+Be^{sx}+\frac{3}{4\pi^2+s}\sin(2\pi x)$$
    finalmente aplicando las condiciones de contorno:
    $$V(0,s)=\int_0^\infty U(0,t)e^{-st}dt=\int_0^\infty 0e^{-st}dt=0$$
    $$V(2,s)=\int_0^\infty U(2,t)e^{-st}dt=\int_0^\infty 0e^{-st}dt=0$$
    de aqui se obtiene:
    $$A=0\;\;\;B=0$$
    Se puede, entonces, escribir la solucion general $V(x,s)$, como sigue:
    $$V(x,s)=\frac{3}{4\pi^2+s}\sin(2\pi x)$$
    <br><br>
    Solo queda aplicar la antitransformada de Laplace:
    $$\mathscr{L}^{-1}\left\{ V \right\}(x,t)=\mathscr{L}^{-1}\left\{\frac{3}{4\pi^2+s}\sin(2\pi x)\right\}$$
    $$\mathscr{L}^{-1}\left\{ V \right\}(x,t)=\sin(2\pi x)\mathscr{L}^{-1}\left\{\frac{3}{4\pi^2+s}\right\}$$
    luego por tabla,<br>
    $$\mathscr{L}^{-1}\left\{ V \right\}(x,t)=3\sin(2\pi x)e^{-4\pi^2 t}$$
    entonces, finalmente:
    <br><br>
    $$U(x,t)=3\sin(2\pi x)e^{-4\pi^2 t}$$

    <b>6)  Hacer un informe te√≥rico con dos aplicaciones a la f√≠sica sobre los Polinomios de Chebyshev de primer y segundo orden.</b><br>

 <div style="text-align: center; background-color:#ffffff; color:#a73434; font-style: normal; font-size:27px; font-family: Time new roman;">
    Polinomios de Chebyshev
</div>

 <div style="text-align: Left; background-color:#ffffff; color:#da6767; font-style: normal; font-size:27px; font-family: Modern;">
    Introducci√≥n
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Los polinomios de Chebychev son dos secuencias polinomiales relacionadas con las funciones coseno y seno, denotados como $T_n(x)$ y $U_n(x)$. Pueden definirse de varias maneras diferentes que llegan al mismo resultado final. Comenzaremos por su definicion trigonometrica:<br>
    Los polinomios de Chebychev de primer tipo $T_n$ son dados por:
    $$T_n(\cos(\theta))=\cos(n\theta)$$
    de manera similar se definen los polinomios de segundo tipo $IU_n$ como:
    $$U_n(\cos(\theta))\sin(\theta)=\sin((n+1)\theta)$$
    Mediante el uso de varias identidades trigonometricas se puede convertir a una forma explisitamente polinomial, por ejemplo para $n=2$ la formula de $T_2$ se puede convertir en un polinomio con el argumento $x=\cos(\theta)$, por medio de la formula de doble angulo:
    $$\cos(2\theta)=2\cos^2(\theta)-1$$
    reemplazando los terminos en la formula con las definiciones anteriores se obtiene:
    $$T_2(x)=2x^2-1$$
    Lo otros $T_n(x)$ se pueden obtener de manera similar.<br>
    Para el caso de los polinomios de segundo tipo usamos la formula de De Moivre para obtener $\sin(n\theta)$ como $\sin(\theta)$ por un polinomio en $\cos(\theta)$ por ejemplo:
    $$\sin(3\theta)=(4\cos^2(\theta)-1)\sin(\theta)$$
    que da como resultado,
    $$U_2(x)=4x^2-1$$
    Una vez convertidos a la forma polinomial $T_n(x)$ y $U_n(x)$ se denominan polinomios de Chebychev polinomios de primer y segundo tipo respectivamente.<br><br>
    Una propiedad importante y conveniente de los polinomios de tipo uno es que son ortogonales con respecto al producto interno respecto a la funcion peso $\frac{1}{\sqrt{1-x^2}}$, definido como sigue:
    $$\left<f,g\right>=\int_{-1}^{1}f(x)g(x)\frac{dx}{\sqrt{1-x^2}}$$
    y analogamente lo son los polinomios de segundo tipo. Esto se sigue del hecho de que los polinomios de Chebychev son solucion de las ecuaciones diferenciales de Chebychev
    $$(1-x^2)y''-xy'+n^2y=0$$
    $$(1-x^2)y''-3xy'+n(n+2)y=0$$
    que son ecuaciones diferenciales de Sturm-Liouville.<br><br>
    Los polinomios de Chebychev presentan gran relevancia en la teoria de aproximaciones ya que las raices de $T_n(x)$, conocidas como nodos de Chebychev, se utilizan como piuntos de coincidencia para utilizar la interpolacion polinomial. El polinomio de interpolacion resultante minimiza el problema del fenomeno de Runge y proporciona una aproximacion cercana a la mejor aproximacion polinomial a una funcion continua.  
</div>

 <div style="text-align: Left; background-color:#ffffff; color:#da6767; font-style: normal; font-size:27px; font-family: Modern;">
    Definicion: Relaciones de Recurrencia
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    La funcion generadora ordinaria para $T_n(x)$ es:
    $$\sum_{n=0}^\infty T_n(x)t^n=\frac{1-tx}{1-2tx+t^2}$$
    Prueba:
:<math>\begin{align}
\text{} \quad  G &\equiv \sum_{n=0}^{\infty}T_n(x)t^n \\
                &= T_0(x) + t T_1(x) + \sum_{n=2}^{\infty}T_n(x)t^n \\
                &= 1 + tx + \sum_{n=0}^{\infty}T_{n+2}(x)t^{n+2} \\
                &= 1 + tx + \sum_{n=0}^{\infty}(2xT_{n+1}(x) - T_n(x))t^{n+2} \\
                &= 1 + tx + \sum_{n=0}^{\infty}2xT_{n+1}(x)t^{n+2} - \sum_{n=0}^{\infty}T_n(x)t^{n+2} \\
                &= 1 + tx + 2tx\sum_{n=0}^{\infty}T_{n+1}(x)t^{n+1} - t^2\sum_{n=0}^{\infty}T_n(x)t^n \\
                &= 1 + tx + 2tx(\sum_{n=0}^{\infty}T_n(x)t^n - 1) - t^2\sum_{n=0}^{\infty}T_n(x)t^n \\
                &= 1 + tx + 2 t x (G - 1) - t^2 G \\  
                &= 1 + tx + 2 t x G - 2 t x - t^2 G \\  
G - 2txG + t^2G &= 1 + t x - 2 t x \\
              G &= \frac{ 1 - t x }{\,1 - 2 t x + t^2\,}
\end{align}</math>
    de lo que se puede obtener:
    :<math>\begin{align}
T_0(x) & = 1 \\
T_1(x) & = x \\
T_{n+1}(x) & = 2 x\,T_n(x) - T_{n-1}(x)~.
\end{align}</math>
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Para el caso de los polinomios de segundo tipo la funcion generadora ordinaria es:
    $$\sum_{n=0}^\infty U_n(x)t^n=\frac{1}{1-2tx+t^2}$$
    de lo que se obtiene:
    :<math>\begin{align}
U_0(x) & = 1 \\
U_1(x) & = 2 x \\
U_{n+1}(x) & = 2 x\,U_n(x) - U_{n-1}(x)~.
\end{align}</math>
</div>

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Productos de polinomios de Chebychev
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Cuando se trabaja con polinomios de Chebychev a menudo se producen productos de ellos los cuales se pueden reducir a combinaciones de polinomios de Chebychev con un grado mas alto o mas bajo y las resoluciones del producto resultan mas convenientes de calcular. Se supondra a continuacion que $m\geq n$ y que $n\geq 0$ para el caso de los poolinomios de primer tipo el producto se expande a:
    $$2T_m(x)T_n(x)=T_{m+n}(x)+T_{|m-n|}(x)$$
    que es una analogia del teorema de la suma 
    $$2\cos(\alpha)\cos(\beta)=\cos(\alpha+\beta)+\cos(\alpha-\beta)$$
    con las identidades:
    $$\alpha=m\arccos(x)\;\; \beta=n\arccos(x)$$
    para $n=1$ esto da como resultado la formula de recurrencia ya conocida y con $n=2$ se forma la relacion de recurrencia para todos los polinomios de Chebychev pares o impares. Se puede entonces concluir tres formulas utiles para evaluar polinomios de Chebychev a partir de esta expansion de productos:
    :<math>\begin{align}
  T_{2n}(x) &= 2\,T_n^2(x)           - T_0(x) &&= 2 T_n^2(x) - 1 \\
T_{2n+1}(x) &= 2\,T_{n+1}(x)\,T_n(x) - T_1(x) &&= 2\,T_{n+1}(x)\,T_n(x) - x \\
T_{2n-1}(x) &= 2\,T_{n-1}(x)\,T_n(x) - T_1(x) &&= 2\,T_{n-1}(x)\,T_n(x) - x 
\end{align}</math>
    para los polinomios de Chebychev de segundo tipo los productos se pueden escribir como:
    $$U_m(x)U_n(x)=\sum_{k=0}^{\infty}U_{m-n+2k}(x)=\sum_{\matrix{p=m-n \\ paso\;2}}^{m+n}U_p(x)$$
    para $m\geq n$.<br> 
</div>

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Relacion entre los dos tipos de los polinomios de Chebychev
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Los polinomios de Chebychev corresponden a un par complem,entario de secuencias de Lucas:
    $$\tilde{V}_n(P,Q) \\ \tilde{U}_n(P,Q)$$ 
    con parametros $P=2x$ y $Q=1$, entonces:
    $$\tilde{U}_n(2x,1)=U_{n-1}(x)$$
    $$\tilde{V}_n(2x,1)=2T_n(x)$$
    luego de ello se deduce que satisfacen un par de ecuaciones de recurrencia mutua:
    $$T_{n+1}(x)=xT_n(x)-(1-x^2)U_{n-1}(x)$$
    $$U_{n+1}=xU_n(x)+T_{n+1}(x)$$
    entonces los polinomios de Chebychev de primer y segundo tipo estan conectados tambien por las siguientes relaciones:
    :<math>\begin{align}
T_n(x) &= \frac{1}{2} \big(U_n(x) - U_{n-2}(x)\big)~. &&\\
T_n(x) &= U_n(x) - x \, U_{n-1}(x)~. &&\\
U_n(x) &= 2\,\sum_{\text{ impar }j}^n T_j(x)      && \text{ impar }n~.\\
U_n(x) &= 2\,\sum_{\text{ par }j}^n T_j(x) - 1 && \text{ par }n~.
\end{align}</math>
    La relacion de recurrencia de la derivada de los polinomios de Chebychev se puede obtener de estas relaciones:
    $$2T_n(x)=\frac{1}{n+1}\frac{d}{dx}T_{n+1}(x)-\frac{1}{n-1}\frac{d}{dx}T_{n-1}(x)$$
    con $n=2,3,4,..$. Esta relacion se utiliza para resolver ecuaciones diferenciales.
    Finalmente nombraremos las relaciones integrales, como sigue:
    $$\int_{-1}^1\frac{T_n(y)}{(y-x)\sqrt{1-y^2}}dy=\pi U_{n-1}(x),$$
    $$\int_{-1}^1\frac{\sqrt{1-y^2}U_{n-1(y)}}{y-x}dy=-\pi T_n(x)$$
</div>

 <div style="text-align: Left; background-color:#ffffff; color:#da6767; font-style: normal; font-size:27px; font-family: Modern;">
    Expresiones Explicitas
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Los diferentes enfoques para definir los polinomios de Chebychev conducen a diferentes expresiones explicitas, tales como:
    :<math>\begin{align}
    T_n(x)  & = \begin{cases}
    \cos(n\arccos x) \qquad & \text{ para }~ |x| \le 1 \\
    \\
     \dfrac{1}{2} \bigg( \Big(x-\sqrt{x^2-1} \Big)^n + \Big(x+\sqrt{x^2-1} \Big)^n \bigg)  \qquad & \text{ para }~ |x| \ge 1 \\
    \end{cases} \\
    \\
     & = \begin{cases}
    \cos(n\arccos x) \qquad \quad & \text{ para }~ -1 \le x \le 1 \\
    \\
    \cosh(n \operatorname{arcosh}x) \qquad \quad & \text{ para }~ 1 \le x \\
    \\
    (-1)^n \cosh\big(n \operatorname{arcosh}(-x)\big) \qquad \quad & \text{ para }~ x \le -1 \\
    \end{cases} \\
    \\
    \\
    T_n(x) & = \sum_{k=0}^{\left \lfloor \frac{n}{2} \right \rfloor} \binom{n}{2k} \left (x^2-1 \right )^k x^{n-2k} \\
    & = x^n \sum_{k=0}^{\left  \lfloor \frac{n}{2} \right \rfloor} \binom{n}{2k} \left (1 - x^{-2} \right )^k \\
    & = \frac{n}{2} \sum_{k=0}^{\left \lfloor \frac{n}{2} \right \rfloor}(-1)^k \frac{(n-k-1)!}{k!(n-2k)!}~(2x)^{n-2k}  \qquad\qquad \text{ for }~ n > 0 \\
    \\
    & = n \sum_{k=0}^{n}(-2)^{k} \frac{(n+k-1)!} {(n-k)!(2k)!}(1 - x)^k \qquad\qquad ~ \text{ para }~ n > 0 \\
    \end{align}</math>
    <math>\begin{align}
    U_n(x) & = \frac{\left (x+\sqrt{x^2-1} \right )^{n+1} - \left (x-\sqrt{x^2-1} \right )^{n+1}}{2\sqrt{x^2-1}} \\
    & = \sum_{k=0}^{\left \lfloor \frac{n}{2} \right \rfloor} \binom{n+1}{2k+1} \left (x^2-1 \right )^k x^{n-2k} \\
    & = x^n \sum_{k=0}^{\left \lfloor \frac{n}{2} \right \rfloor} \binom{n+1}{2k+1} \left (1 - x^{-2} \right )^k \\
    & = \sum_{k=0}^{\left \lfloor \frac{n}{2} \right \rfloor} \binom{2k-(n+1)}{k}~(2x)^{n-2k} & \text{ for }~ n > 0 \\
    & = \sum_{k=0}^{\left \lfloor \frac{n}{2} \right \rfloor} (-1)^k \binom{n-k}{k}~(2x)^{n-2k} & \text{ for }~ n > 0 \\
    & = \sum_{k=0}^{n}(-2)^{k} \frac{(n+k+1)!} {(n-k)!(2k+1)!}(1 - x)^k & \text{ for }~ n > 0 \\
    \end{align}</math>

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Simetria
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    <math>\begin{align}
     T_n(-x) &= (-1)^n T_n(x)
     = \begin{cases}
     T_n(x)  \quad & ~\text{ for }~n~\text{ even} \\
    \\
     -T_n(x) \quad  & ~\text{ for }~n~\text{ odd}
     \end{cases} \\
    \\
    \\
    U_n(-x) &= (-1)^n U_n(x)
    = \begin{cases}
     U_n(x)  \quad & ~\text{ for }~n~\text{ even} \\
    \\
     -U_n(x) \quad & ~\text{ for }~n~\text{ odd}
     \end{cases} \\
    \end{align}</math>
    De lo que se desprende que los polinomios de Chebychev de orden par tienen simetria par y contienen potencias pares de $x$ mientras que los polinomios de orden impar tienen simetria impar y contienen solo potencias impares de $x$.

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Raices y Extremos
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Un polinomio de Chebychev de cualquier tipo con grado $n$ tiene $n$ raices simples diferentes llamadas nodos de Chebychev en el intervalo $[-1,1]$. Usando la definicion trigonometrica y el hecho de que,
    $$\cos((2k+1)\frac{\pi}{2})=0$$
    se puede demostrar entonces que las raices de $T_n(x)$ son:
    $$x_k=\cos\left(\frac{\pi (k+\frac{1}{2})}{n}\right)\;\;k=0,...,n-1$$
    de manera similar las raices de $U_n(x)$ son:
    $$x_k=\cos\left(\frac{k}{n+1}\pi\right)\;\;k=1,...,n$$
    Los extremos de $T_n(x)$ en el intervalo $[-1,1]$ se ubican en:
    $$x_k=\cos\left(\frac{k}{n}\pi\right)\;\;k=0,...,n$$
    Una propiedad unica de los polinomios de Chebychev de primer tipo es que en dicho intervalo todos los extremos tienen valores que son $-1$ o $1$. Tanto el primer como el segundo tipo de polinomios de Chebychev tienen extremos en los puntos finales dados por:
    $$T_n(1)=1$$
    $$T_n(-1)=(-1)^n$$
    $$U_n(1)=n+1$$
    $$U_n(-1)=(-1)^n(n+1)$$
<div>

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Diferenciacion
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Las derivadas de los polinomios tanto de tipo uno como de tipo dos resultan sencillas de realizar y se puede demostrar que
    :<math>\begin{align}
\frac{\mathrm{d}T_n}{\mathrm{d}x} &= n U_{n - 1} \\
\frac{\mathrm{d}U_n}{\mathrm{d}x} &= \frac{(n + 1)T_{n + 1} - x U_n}{x^2 - 1} \\
\frac{\mathrm{d}^2 T_n}{\mathrm{d}x^2} &= n \frac{n T_n - x U_{n - 1}}{x^2 - 1} = n \frac{(n + 1)T_n - U_n}{x^2 - 1}.
\end{align}</math>
    las dos ultimas ecuaciones presentan una indeterminacion especificamente cuando $x=1$ y $x=-1$. Se puede demostrar entonces que:
    <math>\begin{align}
\left. \frac{\mathrm{d}^2 T_n}{\mathrm{d}x^2} \right|_{x = 1} \!\! &= \frac{n^4 - n^2}{3}, \\
\left. \frac{\mathrm{d}^2 T_n}{\mathrm{d}x^2} \right|_{x = -1} \!\! &= (-1)^n \frac{n^4 - n^2}{3}.
\end{align}</math>
    y en general se cumple que:
    $$\frac{d^p}{dx^p}T_n=(\pm 1)^{(n+p)}\prod_{k=0}^{p-1}\frac{n^2-k^2}{2k+1}$$
<div>

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Integracion
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    En cuanto a la integracion la primera derivada de $T_n$ implica que:
    $$\int U_n(x)=\frac{T_{n+1}}{n+1}+C$$
    y la relacion de recurrencia para los polinomios de primer tipo que involucran derivadas establece que para $n\geq 2$,
    $$\int T_n(x)=\frac{1}{2}\left(\frac{T_{n+1}}{n+1}-\frac{T_{n-1}}{n-1}\right)+C$$
    y finalmente se puede probar que:
    $$\int_{-1}^{1}T_n(x)=\left\{\matrix{\frac{-1^n+1}{1-n^2} & n\not=1 \\ 
                                        0                    & n=1}\right.$$
<div>

 <div style="text-align: Left; background-color:#ffffff; color:#da6767; font-style: normal; font-size:27px; font-family: Modern;">
    Ortogonalidad
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'> 
 Tanto $T_n$ como $U_n$ conforman una secuencia de polinomios ortogonales. Para el caso de los polinomios de primer tipo $T_n$ son ortogonales con respecto a la medida,
    $$\frac{1}{\sqrt{1-x^2}}$$
    en el intervalo $[-1,1]$,es decir tenemos:
    $$\int_{-1}^{1}T_n(x)T_m(x)\frac{dx}{\sqrt{1-x^2}}=\left\{\matrix{0 & n \not= m \\ \pi & n=m=0 \\ \frac{\pi}{2} & n=m\not= 0}\right.$$
 <b>Prueba:</b><br> 
    Para probar la igualdad anterior consideramos primeramente $x=\cos(\theta)$, por lo que $dx=-\sin(\theta)d\theta$. Considerando tambi√©n que $T_n(\cos(\theta))=\cos‚Å°(n\theta)$  y $T_m(\cos(\theta))=\cos‚Å°(m\theta)$, reemplazamos en la integral anterior: $$\int_{\pi}^{0}\cos{\left(n\theta\right)\cos{\left(m\theta\right)}}\frac{\left(-sen\left(\theta\right)\right)}{\sqrt{1-\cos^2{\left(\theta\right)}}}d\theta=\int_{0}^{\pi}\cos{\left(n\theta\right)\cos{\left(m\theta\right)}}\frac{sen\left(\theta\right)}{sen\left(\theta\right)}d\theta$$
$$\int_{0}^{\pi}\cos{\left(n\theta\right)\cos{\left(m\theta\right)}}d\theta\;\;(1)$$
A partir de la integral $(1)$ analizamos el caso donde $n=m=0$, y obtenemos:
$$\int_{0}^{\pi}\cos^2{(\theta\ast0)}d\theta=\int_{0}^{\pi}1d\theta=\pi$$
y se demuestra el segundo caso.
Para $n=m\not=0$,   resolviendo la integral $(1)$ nos queda:
    $$\int_{0}^{\pi}\cos^2{(n\theta)}d\theta=\frac{2\pi n+sen(2\pi n)}{4n}=\frac{2\pi n+0}{4n}=\frac{\pi}{2}$$
y se demuestra tercer caso.<br>
    Por √∫ltimo, tomamos el caso donde $n\not=m$ y calculando $(1)$:
 $$\int_{0}^{\pi}\cos{\left(n\theta\right)\cos{\left(m\theta\right)}}d\theta\ =\frac{msen\left(\pi m\right)cos\left(\pi n\right)-ncos\left(\pi m\right)sen\left(\pi n\right)}{m^2-n^2}=\frac{m0cos\left(\pi n\right)-ncos\left(\pi m\right)0}{m^2-n^2}=0$$
Demostrando as√≠ que los polinomios de primer tipo son ortogonales en el intervalo $[-1,1]$.
Para la ortogonalidad de $U_n$ con respecto a la medida,
    $$\sqrt{1-x^2}$$
    en el intervalo $[-1,1]$,es decir tenemos:
    $$\int_{-1}^{1}U_n(x)U_m(x)\sqrt{1-x^2}dx=\left\{\matrix{0 & n \not= m \\ \frac{\pi}{2} & n=m}\right.$$
<b>Prueba:</b><br>
    Tomamos las mismas consideraciones que la prueba anterior, con la diferencia de que ahora $U_n(x)=\frac{sen((n+1)\theta)}{sen(\theta)}$  y $U_m(x)=\frac{sen((m+1)\theta)}{sen(\theta)}$. Reescribiendo la integral:
    $$\int_{\pi}^{0}\frac{sen((n+1)\theta)}{sen(\theta)}\frac{sen((m+1)\theta)}{sen(\theta)}\sqrt{1-\cos^2{(\theta)}}(-sen(\theta))d\theta=\int_{0}^{\pi}\frac{sen((n+1)\theta)}{sen(\theta)}\frac{sen((m+1)\theta)}{sen(\theta)}sen(\theta)sen(\theta)d\theta$$
$$\int_{0}^{\pi}{sen((n+1)\theta)}sen((m+1)\theta)d\theta\;\; (2)$$
De acuerdo con la integral $(2)$ ,analizamos primeramente el caso donde $n=m$, quedando:
$$\int_{0}^{\pi}{sen((n+1)\theta)}sen((n+1)\theta)d\theta=\frac{\pi}{2}-\frac{sen(2\pi n)}{4(n+1)}=\frac{\pi}{2}-0=\frac{\pi}{2}$$
Mientras que para $n‚â†m$:
    $$\int_{0}^{\pi}{sen((n+1)\theta)}sen((m+1)\theta)d\theta=\frac{1}{2}\left(\frac{sen(\pi(m-n))}{m-n}-\frac{sen(\pi(m+n+2))}{m+n+2}\right)=\frac{1}{2}\left(\frac{0}{m-n}-\frac{0}{m+n+2}\right)=0$$
Quedando as√≠ demostrada la ortogonalidad para los polinomios de segundo tipo.

 <div style="text-align: Left; background-color:#ffffff; color:#da6767; font-style: normal; font-size:27px; font-family: Modern;">
    Conjunto base infinito
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
   Para toda funcion del orden de la medida de los polinomios de Chebyshev el conjunto de polinomios de primer tipo forma una base ortonormal de modo que una funcion en dicho espacio en el intervalo $[-1,1]$, puede expresarse mediante la expansion:
    $$f(x)=\sum_{n=0}^{\infty}C_nT_n(x)$$
    como se menciono anteriormente los polinomios de Chebyshev forman una base ortogonal lo que implica que los coeficientes $C_n$ pueden ser determinados facilmente por la aplicacion de un producto interno. Esta suma se llama serie o <i>expansion de Fourier-Chebyshev</i>. Dado que toda serie de Fourier-Chebyshev tiene un equivalente en las series cosenoidales de Fourier por cambio de variables dado en la definicion, se incluyen las siguientes propiedades:
    <ul>
        <li>Los polinomios de Chebyshev forman un conjunto completo.</li>
        <li>La serie de Chebyshev converge a $f(x)$ si la funcion es uniforme y continua por partes.</li>
        <li>Las series de todo funcion es invariante ante adiciones y sustracciones de conjuntos finitos de puntos.</li>
        <li>En una discontinuidad la serie convergera al promedio de los limites derecho e izquierdo.</li>
    </ul>
    La abundacia de teoremas e identidades heredados de las series de Fouriere hace que lso polinomios de Chebyshev sean herramientas importantes en el analisis numerico. 
<div>

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Ejemplo
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
    Consideremos la expansion para $f(x)=\log(1+x)$:
    $$\log(1+x)=\sum_{n=0}^{\infty}C_nT_n(x)$$
    donde los coeficientes $C_n$ pueden ser encontrados mediante la aplicacion de la condicion de ortogonalidad, quedando entonces:
    $$\int_{-1}^1 \frac{dx}{\sqrt{1-x^2}}T_m(x)\log(1+x)=\sum_{n=0}^\infty C_n \int_{-1}^{1}\frac{dx}{\sqrt{1-x^2}}T_m{x}T_n(x)$$
    $$C_n=\left\{\matrix{-\log(2) & n=0 \\ -1^{n+1}\frac{2}{n} & n>0}\right.$$
<div>

 <div style="text-align: Left; background-color:#ffffff; color:#da6767; font-style: normal; font-size:27px; font-family: Modern;">
    Completitud
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
  Ya se ha mencionado la completitud de los polinomios de primer orden dada la completitud de las series cosenoidales de Fourier queda entonces demostrar la completitud del conjunto de los polinomios de segundo orden $U_n(x)$.
    $$\sum_{n=0}^{\infty}U_n(x)U_n(y)=\mu(x)\delta(x-y)$$
    $$\sum_{n=0}^{\infty}\int_{-1}^1dx\sqrt{1-x^2}U_k(x)U_n(x)U_n(y)=\int_{-1}^1dx\sqrt{1-x^2}U_k(x)\mu(x)\delta(x-y)$$
    $$\sum_{n=0}^{\infty}\frac{\pi}{2}\delta_{kn}U_n(y)=\sqrt{1-y^2}U_k(y)\mu(y)$$
    $$\frac{\pi}{2}U_k(y)=\sqrt{1-y^2}U_k(y)\mu(y)$$
    $$\frac{\pi}{2}=\sqrt{1-y^2}\mu(y)$$
    $$\mu(y)=\frac{\pi}{2\sqrt{1-y^2}}$$

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Aplicacion a la aproximacion de funciones respecto de Fourier
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
     Los polinomios de Chebyshev tienen una alta aplicabilidad en aproximaci√≥n de funciones no peri√≥dicas que se definen en un intervalo en espec√≠fico. Aproximando una funci√≥n $f(x)$ con los polinomios de Chebyshev de primer tipo $T_n(x)$, obtenemos la siguiente funci√≥n $P_n(x)$: 
    $$f(x)\simeq P_n(x)=\sum_{n=0}^{N}C_nT_n(x)$$
    Los coeficientes Cn se deben seleccionar de forma tal que $||f(x)-P_n(x)||_2$ sea m√≠nima. Para hallar los $C_n$ partimos de la expresi√≥n anterior, y se multiplica $m.a.m$ por $T_m(x)$ con $m=1,2, 3, ‚Ä¶$:
    $$\sum_{n=0}^{N}{CnTn(x)Tm(x)=f(x)Tm(x)}$$
    Teniendo en cuenta que $n=m=1,2,3,‚Ä¶$ aplicamos la condici√≥n de ortogonalidad en el intervalo $[-1,1]$ quedando:
    $$C_n=\frac{2}{\pi}\int_{-1}^1 \frac{dx}{\sqrt{1-x^2}}f(x)T_n(x)$$
    Un funci√≥n de orden $N$ que utiliza los polinomios de Chebyshev proporciona una aproximaci√≥n precisa para cualquier polinomio de orden $N$ con error cero. Sin embargo, cuando se utiliza la aproximaci√≥n de Fourier requiere de emplear m√°s t√©rminos para alcanzar una precisi√≥n cualitativa comparable. Las siguiente figuras muestran una comparaci√≥n sobre una aproximaci√≥n de Chebyshev con $N=7$ (figura izquierda) para una funci√≥n $f(x)=x^7-14x^5+49x^3-36x$ y otra en donde se realiza la aproximaci√≥n de Fourier para $N=20$.

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
     Para el caso de funciones no polinomiales, sucede que las series de Fourier convergen m√°s r√°pido que los polinomios de Chebyshev. Esto se puede ver tomando como ejemplo una funci√≥n a pasos (Heaviside). (ver Figura 2).

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
     Para aproximar la funci√≥n escal√≥n mediante Chebyshev se necesitan $N=20$ mientras que el caso de Fourier se requiere de √∫nicamente $10$ t√©rminos para obtener una precisi√≥n cualitativa comparable. Esto se debe a que la funci√≥n tomada se corresponde a una funci√≥n impar y los $n$ correspondiente a $C_n$ son pares e impares, por lo que al realizar la integral en el intervalo $[-1,1]$ del producto de $f(x)T_n(x)$ con respecto a la medida $\frac{1}{\sqrt{1-x^2}}$ nos quedaran √∫nicamente aquellos t√©rminos asociados a n par puesto que el producto de una funci√≥n para por otra impar es cero. Esto indica que de los $N=20$ de la aproximaci√≥n mediante polinomios Chebyshev, solo $10$ contribuyen a la aproximaci√≥n, es decir los t√©rminos impares de los polinomios de Chebyshev.

 <div style="text-align: Left; background-color:#ffffff; color:##555555; font-style: normal; font-size:27px; font-family: Modern;">
    Solucion a la ecuacion de Laplace en coordenadas elipticas en funcion de los polinomios de Chebyshev
</div>

<div style='text-align:justify; background-color:#ffffff; color:#000000; font-style: normal; font-size:22px; font-family: Modern'>
     Considerando el Laplaciano d ela ecuacion en coordenadas elipticas $(\mu,\nu)$:
    $$\nabla^2 u(\mu,\nu)=\frac{1}{C^2\left(\sinh^2(\mu)+\sin^2(\nu)\right)}\left(\partial^2_{\mu}+\partial^2_{\nu}\right)u(\mu,\nu)=0$$
    es evidente que para el caso homogeneo se convierte en:
    $$\left(\partial^2_{\mu}+\partial^2_{\nu}\right)u(\mu,\nu)=0$$
    proponiendo una solucion por separacion de variables:
    $$u(\mu,\nu)=M(\mu)N(\nu)$$
    entonces:
    $$M''-\alpha M=0$$
    $$N''+\alpha N=0$$
    es sabido que el primero tiene soluciones senoidales y cosenoidales, mientras que la segunda ecuacion tiene soluciones hiupervolicas. 
    $$u_n=\left(A_n\cosh(n\mu)+B_n\sinh(n\mu)\right)\left(C_n\cos(n\nu)+D_n\sin(n\nu)\right)$$
    pudiendo reescribir la solucion anterior en terminos d elso polinomios de Cjhebyshev, como sigue:
    $$u_n=\left(a_n\cosh(n\mu)+b_n\sinh(n\mu)\right)\left(c_nT_n(x)+d_nU_{n-1}(x)\sqrt{1-x^2}\right)$$
    con $x=\cos(\nu)$.
    $$u_n=\left(a_n\cosh(n\mu)+b_n\sinh(n\mu)\right)\left(c_nT_n(x)+d_n\left(2xU_{n}(x)-U_{n+1}(x)\right)\sqrt{1-x^2}\right)$$
