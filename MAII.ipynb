Sin perder generalidad $l\ge k$ e integrando por partes $l+m$ veces, donde:

$$u = \pmatrix {1 - x^2}^m \frac {d^{k + m} } {d x^{k + m} } \pmatrix {\pmatrix {x^2 - 1}^k}$$

$$\frac {\pmatrix {-1}^{l + m} } {2^{k + l} k! l!} \int_{-1}^1 \pmatrix {x^2 - 1}^l \frac {d^{l + m} } {d x^{l + m} } \pmatrix {\pmatrix {1 - x^2}^m \frac {d^{k + m} } {d x^{k + m} } \pmatrix {\pmatrix {x^2 - 1}^k} } d x$$

Expandiendo el segundo factor por la regla de Leibnitz: 

$$\frac {d^{l + m} } {d x^{l + m} } \pmatrix {\pmatrix {1 - x^2}^m \frac {d^{k + m} } {d x^{k + m} } \pmatrix {\pmatrix {x^2 - 1}^k} } = \sum_{r \mathop = 0}^{l + m} \binom {l + m} r \frac {d^r} {d x^r} \pmatrix {\pmatrix {1 - x^2}^m} \frac {d^{l + k + 2 m - r} } {d x^{l + k + 2 m - r} } \pmatrix {\pmatrix {x^2 - 1}^k}$$

Recordando la condicion de que $m\leq l$ la primera derivada en la sumatoria sera distinta de $0$ si $r\leq 2m$ mientras que la segunda derivada  sera distinta de $0$ si $k + l + 2 m - r \le 2 k$ lo que implica $r \ge 2 m + l - k$.

Dado que $l \ge k$, las condiciones anteriores implican que los terminos destintos de cero en la sumatoria ocurren cuando $r=2m$ y $l=k$, quedando:

$$\pmatrix {-1}^l \delta_{k l} \frac {\pmatrix {-1}^{l + m} } {2^{2 l} \pmatrix {l!}^2} \binom {l + m} {2 m} \int_{-1}^1 \pmatrix {x^2 - 1}^l \frac {d^{2 m} } {d x^{2 m} } \pmatrix {\pmatrix {1 - x^2}^m} \frac {d^{2 l} } {d x^{2 l} } \pmatrix {\pmatrix {1 - x^2}^l}  d x$$

Donde $\delta_{k l}$ es la Delta de Kronecker que muestra la ortogonalidad de estas funciones cuando $l \ne k$.

### La ecuacion diferencial para las funciones de Hermite sera:

$$y''_n-x^2y_n=-(2n+1)y_n\;\; n=1,2,3,$$

Se resolvera utilizando el operador $D\equiv \frac{d}{dx}$.

$$(D-x)(D+x)y=(D-x)(y'+xy)=y''-x^2+y$$
y de forma similar:
$$(D+x)(D-x)y=y''-x^2-y$$

De esta manera podemos escribir la ecuacion diferencial como:

$$(D-x)(D+x)y_n=-2ny_n$$
$$(D+x)(D-x)y_n=-2(n+1)y_n$$

luego podemos operar sobre estas como sigue:
$$(D-x)(D+x)[(D+x)y_m]=-2m[(D+x)y_m]$$
$$(D+x)(D-x)[(D-x)y_m]=-2(m+1)[(D-x)y_m]$$

podemos ver que si $n=m\pm 1$ y $(D+x)y_m=y_n$ o $(D-x)y_m=y_n$:

## $$y_{m-1=}(D+x)y_m\hspace{1cm} y_{m+1=}(D-x)y_m$$
